# Audit-Closed AI Scientist Results

This notebook loads and inspects `results/experiment_results.json` generated by:

```bash
python run_all_experiments.py
```

````python
import json
from pathlib import Path

results_path = Path('results/experiment_results.json')
results = json.loads(results_path.read_text(encoding='utf-8'))
sorted(results.keys())
````

````python
benchmark = results['benchmark']
summary = {
    'false_discovery_rate': benchmark['false_discovery_rate'],
    'replicability': benchmark['replicability'],
    'evidence_stability_under_sequential_testing': benchmark['evidence_stability_under_sequential_testing'],
    'robustness_to_adversarial_experiments': benchmark['robustness_to_adversarial_experiments'],
}
summary
````

````python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

figure_files = [
    'figures/false_discoveries_vs_hypotheses.png',
    'figures/candidate_shopping_false_discovery.png',
    'figures/p_vs_e_reliability.png',
    'figures/power_curve_detection.png',
    'figures/sentinel_hierarchy_stress.png',
    'figures/drift_localization_uptime.png',
    'figures/replication_success_rate.png',
    'figures/adversarial_robustness.png',
]

for path in figure_files:
    img = mpimg.imread(path)
    plt.figure(figsize=(8, 4.5))
    plt.imshow(img)
    plt.axis('off')
    plt.title(path)
    plt.show()
````
